```python
# Defining the non-linear model for fitting using the square root function
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Preparing the data for linear regression
X = np.array([10, 15, 20, 25, 30, 35, 40, 45, 50])  # Discount in %
Y = np.array([3.05, 7.62, 12.19, 20.42, 28.65, 42.06, 55.47, 74.68, 93.88])  # Increase in sale in %

def nonlinear_sqrt_model(x, a, b):
    return a * np.sqrt(x) + b

# Performing the non-linear regression using the square root model
popt_sqrt, pcov_sqrt = curve_fit(nonlinear_sqrt_model, X, Y)

# Plotting the data and the non-linear fit
plt.figure(figsize=[10,6])
plt.scatter(X, Y, color='red', label='Data points')
x_values_sqrt = np.linspace(min(X), max(X), 100)
y_values_sqrt = nonlinear_sqrt_model(x_values_sqrt, *popt_sqrt)
plt.plot(x_values_sqrt, y_values_sqrt, label='Non-linear (sqrt) fit', color='blue')

plt.xlabel('Product Discount in %')
plt.ylabel('Increase in sale in %')
plt.title('Non-linear Regression with Square-root Model')
plt.legend()
plt.show()

# Returning the parameters of the fit
popt_sqrt
```

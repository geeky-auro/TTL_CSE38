## Groupby allows you to group together rows based off of a column and perform an aggregate function on them.

```python
import numpy as np
import pandas as pd
```

```python
data={'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
      'Person':['Sam','Charlie','Amy','Vamessa','Carl','Sarah'],
      'Sales':[200,120,340,124,243,350]}
```

```python
df=pd.DataFrame(data)
df
```

```python
byComp=df.groupby('Company')
```

```python
print((byComp['Sales'].mean()))
```

```python
print(byComp['Sales'].sum().loc['FB'])
```

```python
# One Liner Code for Group By
df.groupby('Company').count().loc['FB']
```

```python
df.groupby('Company').describe().transpose()['FB']
# DataFrame Structure: If your DataFrame df does not have a 'Company' column, or if 'FB' is not a valid index label, you would encounter errors.
# Data Content: If there is no data associated with the 'FB' company in your DataFrame, or if there are missing values that affect the calculations, you might not get the desired output.
# df.groupby('Company').describe().transpose().loc['FB']
```

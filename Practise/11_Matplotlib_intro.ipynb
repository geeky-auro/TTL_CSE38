````python
import matplotlib.pyplot as plt
````

````python
import numpy as np
# Return evenly spaced numbers over a specified interval.
x=np.linspace(0,5,11)
y=x**2
````

````python
x
````

````python
y
````

````python
plt.plot(x,y,'g-')
````

````python
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
plt.plot(x,y,'r-')
````

````python
# Add an Axes to the current figure or retrieve an existing Axes.
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
````

````python
# object oriented Method 
# This is an imaginary Canvas
fig=plt.figure()

# Now we need to add axes 
# About the first Parameter 
# This parameter is the dimensions [left, bottom, width, height] of the new axes.
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y,'r')
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
````

````python
fig=plt.figure()
# Note the parameteric Values must be between 0's and 1's 
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
# Basically the plots dimensions give us the idea of how far from the x axis,y-axis the distance should be and   
# 3rd and 4th Parameter give us the idea of what is the size of the plot i.e (width,height) 
axes2=fig.add_axes([0.4,0.2,0.4,0.3])
axes1.plot(y,x)
axes1.set_title('Larger Plot')
axes2.plot(x,y)
axes2.set_title('Smaller Plot')
````

# Summary of 1st Part
- Step 1: Create a Empty Canvas ;) -> plt.figure()
- Step 2: Create a Graph by giving dimensions ;) -> ```plt.figure().add_axes([left_dist,bottom_dist,width,height])```
- Step 3: Plot Your Readings to the Graph ;) -> ```plt.figure().add_axes(Necessary Parameters).plot(x_axis,y_axis)```

````python
# Where Subplots differ is we can specify Number of Rows and Columns 
fig,axes=plt.subplots(nrows=3,ncols=3)
# This might give a design issues of overlapping figure
plt.tight_layout()
# axes.plot(x,y)
````

````python
fig,axes=plt.subplots(nrows=1,ncols=2)
axes[0].plot(x,y),axes[1].plot(x,y)
axes[0].set_title('First Plot')
axes[1].set_title('Second Plot|')
````

````python
fig,axes=plt.subplots(nrows=1,ncols=4)
plt.tight_layout()
````

# Figure Size and DPI

````python
# The tuple we pass in is the Fig Suze width and height of the argument 
# Dpi is nothing but the dots oer inch or pixels per inch
# Note DPI is optional we can just ignore ;0
fig=plt.figure(figsize=(8,2),dpi=100)

ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
````

````python
# the above can also be done with subplots
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
````

````python
# How to save a Figure to file

fig.savefig('My_picture.png',dpi=200) 
````

````python
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('Title')
ax.set_ylabel('Y')
ax.set_xlabel('X')
````

````python
# Legends -> label text to show what plot is what plot
fig=plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.set_title('Title')
ax.set_ylabel('Y')
ax.set_xlabel('X')
ax.plot(x**2,y,"r",label='X Square')
ax.plot(y**3,x,"g",label='Y Cubed')

ax.legend()
````

````python
# Sometimes legend can also overlap ... -)
# it is called as location code 

fig=plt.figure()

ax=fig.add_axes([0,0,1,1])

ax.set_title('Title')
ax.set_ylabel('Y')
ax.set_xlabel('X')
ax.plot(x**2,y,"r",label='X Square')
ax.plot(y**3,x,"g",label='Y Cubed')

ax.legend(loc=10)
````

# Plot Appearance 

````python
# Settign colours with Matplotlib

fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple')
````

````python
# To edit line
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',linewidth=3,alpha=0.6) 
# Alpha tells us the transparency of the code ;)
ax.plot(x,y,color='purple',lw=3,alpha=0.6) 
````

````python
# We can also style the line
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
# ax.plot(x,y,color='purple',lw=3,linestyle='--')
# ax.plot(x,y,color='purple',lw=3,linestyle='-.')
ax.plot(x,y,color='purple',lw=3,ls='--')
````

````python
# use of Markers
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y,color='purple',marker='*',markersize=10,markerfacecolor='yellow')
````

```python
import seaborn as sns
tips=sns.load_dataset('tips')
tips.head()
```

```python
# Bar Plots
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.mean)
# It ususally considers Means ;)
# Estimator parameter usually Takes different function parameter for mapping such as we can give standard deviation or mean!
```

```python
sns.boxplot(x='day',y='total_bill',data=tips)
```

Box plots, violin plots, and swarm plots are considered advanced data visualization techniques because they provide more detailed and nuanced representations of the distribution of data compared to simpler plots like bar charts or line graphs. These advanced techniques are needed for several reasons:

- **Revealing distribution shape and characteristics:**
Simple plots only show summary statistics like means or totals
Advanced plots reveal the full shape of the distribution (skewness, modality, outliers, etc.)
This is crucial for understanding the true nature of the data beyond just central tendencies
- **Comparing distributions across groups:**
These plots make it easy to compare distributions side-by-side across different categories or groups
This allows for identifying similarities, differences, and patterns in how variables distribute across groups
- **Dealing with overlapping/dense data:**
For datasets with many observations or overlapping points, simple scatter plots become over-plotted and unclear
Techniques like swarm plots and violin plots handle this denseness in a clean and interpretable way
- **Multivariate data exploration:**
While simple plots work for one or two variables, these techniques facilitate exploring relationships between a continuous variable and multiple categorical variables simultaneously
- **Communicating nuanced insights:**
The rich visualizations enable communicating complex characteristics and patterns in the data distribution that would be difficult to convey otherwise
- **Supporting statistical analyses:**
Many statistical techniques rely on understanding the full data distribution
These plots support assessing if the necessary distribution assumptions are met
In summary, while simple plots are useful for high-level summaries, advanced techniques like box plots, violin plots, and swarm plots are crucial for gaining deeper insights into the nuances of data distributions, making comparisons across groups, and supporting more rigorous statistical analyses.

```python
sns.boxplot(x='day',y='total_bill',data=tips,hue='smoker')
```

```python
# Violin Plot
sns.violinplot(x='day',y='total_bill',data=tips)
# A violin plot is a hybrid of a box plot and a kernel density plot, which shows peaks in the data. 
# It is used to visualize the distribution of numerical data. 
# Unlike a box plot that can only show summary statistics, violin plots depict summary statistics and the density of each variable.
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,hue='sex',split=True)
```

```python
#Strip plot
sns.stripplot(x='day',y='total_bill',data=tips,jitter=False)
```

```python
# To get to know how many points are stacked to each other ..!
# use the parameter Jitter ;)
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
```

 # Swarm Plots
 Swarmplots are visualization technique that provides an alternative to traditional boxplots or violin plots for displaying the distribution of categorical data. The main advantage of using swarmplots is that they can better reveal the underlying structure and density of the data, especially when there are many data points or when the data is highly overlapping or concentrated in certain areas.
- Avoids overplotting/overlapping of data points
- Clearly reveals data density within categories
- Easily identifies outliers
- Can reveal multimodal distributions within categories
- Facilitates visual comparison of distributions across categories
- Handles large datasets effectively by adjusting data point positions
- Complements other visualization techniques like boxplots and violin plots

```python
# A swarmplot is a type of categorical scatter plot used to visualize the distribution of data points in a dataset. It is a useful tool for exploring and analyzing the relationship between two or more variables in a dataset.
sns.swarmplot(x='day',y='total_bill',data=tips)
# visualization technique that provides an alternative to traditional boxplots or violin plots for displaying the distribution of categorical data. 
#  The main advantage of using swarmplots is that they can better reveal the underlying structure and density of the data, especially when there are many data points or when the data is highly overlapping or concentrated in certain areas.
```

```python
sns.violinplot(x='day',y='total_bill',data=tips)
sns.swarmplot(x='day',y='total_bill',data=tips,color='black')
```

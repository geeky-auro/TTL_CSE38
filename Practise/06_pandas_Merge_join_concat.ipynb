# Merging
The Merge function allows you to merge DataFrames together using a similar Logic as Merging SQL Tables together. For Example

```python
# Perform Inner Join Operation
pd.merge(left=left,right=right,on=['Key','Key'],how='inner')
```

```python
left=pd.DataFrame({
    'Key1':['k0','k1','k2','k2'],
    'Key2':['k0','k1','k0','k1'],
    'A':['A0','A1','A2','A3'],
    'B':['B0','B1','B2','B3']
})


right=pd.DataFrame({
    'Key1':['k0','k1','k1','k2'],
    'Key2':['k0','k0','k0','k0'],
    'C':['C0','C1','C2','C3'],
    'D':['D0','D1','D2','D3']
})
```

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
    
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                      'key2': ['K0', 'K0', 'K0', 'K0'],
                         'C': ['C0', 'C1', 'C2', 'C3'],
                         'D': ['D0', 'D1', 'D2', 'D3']})
pd.merge(left,right,on=['key1','key2'])
```

**All keys from both dataframes will be interpreted like a tuple.** If you refer to the Jupyter Notebook for this lecture,
~~~
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
    
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                      'key2': ['K0', 'K0', 'K0', 'K0'],
                         'C': ['C0', 'C1', 'C2', 'C3'],
                         'D': ['D0', 'D1', 'D2', 'D3']})
~~~
Here the keysets that will be interpreted as tuples will be:

**(K0,K0), (K0,K1), (K1,K0) and (K2,K1)** from left and

**(K0,K0), (K1,K0), (K1,K0) and (K2,K0)** from right.

Now inner causes intersection and outer causes union(you must be familiar with set theory). inner will cause merging of only those values for which the corresponding tuples match.

Here first tuple (K0,K0) from left matches with the first tuple (K0,K0) from right. So those values are reflected in the output.

0 A0 B0 C0 D0

Second tuple (K0,K1) from left has no match in right, so it has been excluded. The third tuple (K1,K0) left has two matches in right so it appears twice with the corresponding values.

1 A2 B2 C1 D1

2 A2 B2 C2 D2

And the fouth tuple (K2,K1) left again has no match right so it doesn't shows up.



outeron the othe hand merges the dataframes even if the keys dont match and insets 'NaN' for the values corresponding to keys that dont match.

Notice how the second tuple (K0,K1) from left which had no matching tuple in right is shown with 'NaN'. Similarly fouth tuple (K2,K1) from left and fouth tuple (K2,K0) from right bears 'NaN' as values.

0 K0 K0 A0 B0 C0 D0

1 K0 K1 A1 B1 NaN NaN

2 K1 K0 A2 B2 C1 D1

3 K1 K0 A2 B2 C2 D2

4 K2 K1 A3 B3 NaN NaN

5 K2 K0 NaN NaN C3 D3

```python
pd.merge(left=left,right=right,on=['key1','key2'],how='outer')
```

| LEFT JOIN                                                          | RIGHT JOIN                                                         |
|--------------------------------------------------------------------|--------------------------------------------------------------------|
| It joins two or more tables, returns all records from the left table, and matching rows from the right-hand table. | It is used to join two or more tables, returns all records from the right table, and matching rows from the left-hand table. |
| The result-set will contain null value if there is no matching row on the right side table. | The result-set will contain null value if there is no matching row on the left side table. |
| It is also known as LEFT OUTER JOIN.                              | It is also called as RIGHT OUTER JOIN.                             |


```python
pd.merge(left=left,right=right,on=['key1','key2'],how='left')
```

```python
pd.merge(left=left,right=right,on=['key1','key2'],how='right')
```

# Joining
Joining is a convenient method for combining the columns of two potentially differently-indexed DataFrames into a single result
DataFrame.

```python
left=pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
'B': ['B0', 'B1', 'B2', 'B3']},index=['k0','k1','k0','k3'])


right=pd.DataFrame({'C': ['C0', 'C1', 'C2', 'C3'],
'D': ['D0', 'D1', 'D2', 'D3']},index=['k0','k2','k2','k3'])
```

```python
left.join(right)
```

```python
right.join(left)
```

```python
right.join(left,how='outer')
```

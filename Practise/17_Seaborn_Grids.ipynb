```python
import seaborn as sns
import matplotlib.pyplot as plt
iris=sns.load_dataset('iris')
iris.head()
```

```python
# ABout the Data
iris['species'].unique()
# Other 4 Columns are the feature of the data
```

```python
# Automated Joint Plot for the data
g=sns.pairplot(iris)
g.map(plt.scatter)
```

```python
g=sns.PairGrid(iris)
g.map(plt.scatter)
```

```python
g.map_diag(sns.displot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

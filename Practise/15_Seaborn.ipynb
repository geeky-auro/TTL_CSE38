```python
#Distribution plot
import seaborn as sns
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# dist Plot

# Dist plot helps us to plot Observation of a univairate Variable!
# It is a univariant set of collected data, which means the data distribution of one variable will be shown against another variable.
sns.displot(tips['total_bill'],kde=True,bins=40)
# A kernel density estimate (KDE) plot is a method for visualizing the distribution of observations in a dataset, analogous to a histogram. KDE represents the data using a continuous probability density curve in one or more dimensions
# The bins parameter enables you to control the bins of the histogram (i.e., the number of bars)
```

```python
# Joint Plot
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
# TO change the style of the Scatterplot there's something parameter called as kind
```

The joint plot is a way of understanding the relationship between two variables and the distribution of individuals of each variable. The joint plot mainly consists of three separate plots in which, one of it was the middle figure that is used to see the relationship between x and y.

```python
sns.jointplot(x='total_bill',y='tip',data=tips,)
```

A pair plot, also known as a scatterplot matrix, is a matrix of graphs that shows the relationship between variables in a dataset.

```python
# Pair Plot ;)
# Pair Plot will do Joint plot for every possible columns in a numerical dataframe
# Hue is used for categorical values such as male and female or gender,etc
sns.pairplot(tips,hue='sex')
```

```python
sns.pairplot(tips,hue='sex',palette='coolwarm')
```

```python
# Rug Plot -> Draws a Dash mark for every Single Univariate Point (Single Column)!
sns.rugplot(tips['total_bill'])
```

```python
sns.displot(tips['total_bill'])
# Compare between displot and rug plot the more high bar is the more number of dashes in the particular area ;)
```

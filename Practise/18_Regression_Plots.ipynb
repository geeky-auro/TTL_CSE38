```python
import seaborn as sns
```

```python
%matplotlib inline
```

# Lmplot stands for Linear Model Plot

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex')
#You can use the hue parameter when creating pairplots in seaborn to color plot aspects based on the values of a specific variable
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],
           scatter_kws={'s':100})
# scatter_kws is used to increase the size of markers ;)
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips)
```

```python
# Instead of hue we can use Cols
# hue was used to show extra attributes in scatter plot over a categorical data
# You can use the hue parameter when creating pairplots in seaborn to color plot aspects based on the values of a specific variable
sns.lmplot(x='total_bill',y='tip',data=tips,col='sex',hue='day')
```
